```python
#optimized
A = [5,4,0,3,1,6,2]
res=0
seen=[False]*len(A)
for i in range(len(A)):
    step=0
    while not seen[i]:
        seen[i]=True           
        i=A[i]
        step+=1
    if step>res:
        res=step

res
       
        
```
